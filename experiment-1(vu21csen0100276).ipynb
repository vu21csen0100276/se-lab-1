```python
#Hardcoded Variables
import matplotlib.pyplot as plt
import numpy as np

a = 1
b = 3
c = -6

def weather_model(x):
    return a*x**2 + b*x + c

x = np.linspace(-10, 10, 400)
y = weather_model(x)

plt.plot(x, y)
plt.title('Weather Model')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

with open('variables.txt', 'r') as file:
    a, b, c = map(float, file.readline().split())

def weather_model(x):
    return a*x**2 + b*x + c

x = np.linspace(9, 10, 100)
y = weather_model(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label=f'{a}x^2 + {b}x + {c}')
plt.title('Plot of the quadratic function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Multiple sets of inputs
sets = [(2, 4, 6), (1, 3, 5), (3, 7, 9)]

def weather_model(x, a, b, c):
    return a*x**2 + b*x + c

x = np.linspace(9, 10, 100)

for a, b, c in sets:
    y = weather_model(x, a, b, c)
    plt.plot(x, y, label=f'a={a}, b={b}, c={c}')

plt.title('Weather Model')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Hardcoded variables
a = 4
b = 2
c = 7

def weather_model(x, a, b, c):
    return a*x**2 + b*x + c

x = np.linspace(9, 10, 100)
y = weather_model(x, a, b, c)

plt.plot(x, y, label=f'Hardcoded: a={a}, b={b}, c={c}')

# Variables input from keyboard
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

y = weather_model(x, a, b, c)

plt.plot(x, y, label=f'Input: a={a}, b={b}, c={c}')

plt.title('Weather Model')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.grid(True)
plt.legend()
plt.show()
```
